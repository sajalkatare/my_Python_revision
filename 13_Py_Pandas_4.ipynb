```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading Titanic Dataset

#In below filew sepertor is tab

titanic_train = pd.read_csv('https://gist.githubusercontent.com/michhar/2dfd2de0d4f8727f873422c5d959fff5/raw/ff414a1bcfcba32481e4d4e8db578e55872a2ca1/titanic.csv', sep = '\t')

titanic_train
```

```python
#drop all the columns where Nan is present
titanic_train = titanic_train.dropna(axis = 1)
```

```python
titanic_train
```

### Apply Function

#### Below code will add a new column to data set which will have length of Name column

```python
titanic_train['Name_len'] = titanic_train['Name'].apply(len)
```

```python
titanic_train
```

#### Below code will give square root of Fare column


```python
titanic_train['sq_rt_fare2'] = titanic_train['Fare'].apply(lambda x: np.sqrt(x))
```

```python
titanic_train['sq_rt_fare'] = titanic_train['Fare'].apply(np.sqrt)
```

```python
titanic_train
```

### Using User Defined Function in Apply

```python
def test1(x):
    if x == 'female':
        return 0
    else:
        return 1
```

```python
titanic_train['flag'] = titanic_train['Sex'].apply(test1)
```

```python
titanic_train
```
